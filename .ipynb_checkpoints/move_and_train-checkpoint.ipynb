```python
#Modified script to split data in between PATHOLOGICAL and NORMAL
#Slight error - the variable new_extension is not concatenated in sink_loc IE images are not converted to JPEGs.
import os
import csv
import pandas as pd
from shutil import copyfile
from PIL import Image


data_loc = '/content/.kaggle/datasets/nih-chest-xrays/data/Data_Entry_2017.csv'
files_size = 100
#Gets a list of all the classifications present, and adds them to the variable classifications, which is returned.

labels_data = pd.read_csv(data_loc)
normal_idx = []
pathological_idx = []
for i in range(len(labels_data)):
  if labels_data.iloc[i,1].split('|')[0] == "No Finding":
    normal_idx.append(labels_data.iloc[i,0].split('|')[0])
  else:
    pathological_idx.append(labels_data.iloc[i,0].split('|')[0])

print ("Normal:", len(normal_idx), "Pathological:", len(pathological_idx))

source_folder = '/content/images/'

len_normal_idx = len(normal_idx)
len_pathological_idx = len(pathological_idx)

halfway_normal_idx = int((len(normal_idx))/2)
halfway_pathological_idx = int((len(pathological_idx))/2)

for i in range(0, halfway_normal_idx):
  source_loc = source_folder + normal_idx[i]
  im = Image.open(source_loc)
  im = im.convert('RGB')
  new_extension = normal_idx[i].split('.')[0] + '.jpg'
  sink_loc = '/content/image_dataset/train/normal/' + normal_idx[i]
  copyfile(source_loc, sink_loc)

print ("Train-Normal Complete...")

for i in range(halfway_normal_idx, len_normal_idx):
  source_loc = source_folder + normal_idx[i]
  im = Image.open(source_loc)
  im = im.convert('RGB')
  new_extension = normal_idx[i].split('.')[0] + '.jpg'
  sink_loc = '/content/image_dataset/validation/normal/' + normal_idx[i]
  copyfile(source_loc, sink_loc)

print ("Validation-Normal Complete...")

for i in range(0, halfway_pathological_idx):
  source_loc = source_folder + pathological_idx[i]
  im = Image.open(source_loc)
  im = im.convert('RGB')
  new_extension = pathological_idx[i].split('.')[0] + '.jpg'
  sink_loc = '/content/image_dataset/train/pathological/' + pathological_idx[i]
  copyfile(source_loc, sink_loc)

print ("Train-Pathological Complete...")

for i in range(halfway_pathological_idx, len_pathological_idx):
  source_loc = source_folder + pathological_idx[i]
  im = Image.open(source_loc)
  im = im.convert('RGB')
  new_extension = pathological_idx[i].split('.')[0] + '.jpg'
  sink_loc = '/content/image_dataset/validation/pathological/' + pathological_idx[i]
  copyfile(source_loc, sink_loc)

print ("Validation-Pathological Complete...")
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K


# dimensions of our images.
img_width, img_height = 1024, 1024

train_data_dir = '/content/image_dataset/train'
validation_data_dir = '/content/image_dataset/validation'
nb_train_samples = 2000
nb_validation_samples = 800
epochs = 50
batch_size = 16

if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

# this is the augmentation configuration we will use for testing:
# only rescaling
test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)

model.save_weights('first_try.h5')
```
